```python
import yfinance as yf
import warnings
warnings.filterwarnings("ignore")
```

```python
ticker = yf.Ticker("COCA34.SA") 
info = ticker.info
```

```python
def get_data():
    return ticker.history(period="max", interval="1d")
```

```python
historical_data = get_data()
```

```python
historical_data.reset_index(inplace=True)
```

```python
historical_data.info()
```

```python
historical_data
```

```python
historical_data["Day"] = historical_data["Date"].dt.day
historical_data["Month"] = historical_data["Date"].dt.month
historical_data["Year"] = historical_data["Date"].dt.year
historical_data["Weekday"] = historical_data["Date"].dt.day_name()
```
